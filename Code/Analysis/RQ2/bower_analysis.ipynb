```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV files into DataFrames
adaptable_df = pd.read_csv('../../../Datasets/RQ2/bower/bower_adaptable_issues.csv')
consistent_df = pd.read_csv('../../../Datasets/RQ2/bower/bower_consistent_issues.csv')
intentional_df = pd.read_csv('../../../Datasets/RQ2/bower/bower_intentional_issues.csv')
responsible_df = pd.read_csv('../../../Datasets/RQ2/bower/bower_responsible_issues.csv')

# Combine DataFrames
all_data = pd.concat([adaptable_df, consistent_df, intentional_df, responsible_df])

# Convert 'version' column to datetime
all_data['version'] = pd.to_datetime(all_data['version'], errors='coerce')

# Drop rows with missing or invalid dates
all_data = all_data.dropna(subset=['version'])

# Sort DataFrame by 'version'
all_data = all_data.sort_values(by='version')

# Explore the data
print(all_data.head())
print(all_data.describe())

```

```python
plt.figure(figsize=(10, 6))
plt.plot(all_data['version'], all_data['total_debt_high'], label='High')
plt.plot(all_data['version'], all_data['total_debt_medium'], label='Medium')
plt.plot(all_data['version'], all_data['total_debt_low'], label='Low')

plt.title('Total Technical Debt Over Versions')
plt.xlabel('Version')
plt.ylabel('Total Debt')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(all_data['version'], all_data['security_issues_high'], label='Security (High)')
plt.plot(all_data['version'], all_data['reliability_issues_high'], label='Reliability (High)')
plt.plot(all_data['version'], all_data['maintainability_issues_high'], label='Maintainability (High)')

plt.title('Code Smell Issues (High Severity) Over Versions')
plt.xlabel('Version')
plt.ylabel('Number of Issues')
plt.legend()
plt.show()
```
