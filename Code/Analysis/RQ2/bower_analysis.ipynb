```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV files into DataFrames
adaptable_df = pd.read_csv('../../../Datasets/RQ2/bower/bower_adaptable_issues.csv')
consistent_df = pd.read_csv('../../../Datasets/RQ2/bower/bower_consistent_issues.csv')
intentional_df = pd.read_csv('../../../Datasets/RQ2/bower/bower_intentional_issues.csv')
responsible_df = pd.read_csv('../../../Datasets/RQ2/bower/bower_responsible_issues.csv')

# Combine DataFrames
all_data = pd.concat([adaptable_df, consistent_df, intentional_df, responsible_df])

# Convert 'version' column to datetime
all_data['version'] = pd.to_datetime(all_data['version'], errors='coerce')

# Drop rows with missing or invalid dates
all_data = all_data.dropna(subset=['version'])

# Sort DataFrame by 'version'
all_data = all_data.sort_values(by='version')

# Explore the data
print(all_data.head())
# print(all_data.describe())

```

```python
# Total count for each severity level
total_low = all_data['security_issues_low'] + all_data['reliability_issues_low'] + all_data['maintainability_issues_low']
total_medium = all_data['security_issues_medium'] + all_data['reliability_issues_medium'] + all_data['maintainability_issues_medium']
total_high = all_data['security_issues_high'] + all_data['reliability_issues_high'] + all_data['maintainability_issues_high']

# Print total count for each column
print(f'Total Count for Security Issues (Low): {all_data["security_issues_low"].sum()}')
print(f'Total Count for Reliability Issues (Low): {all_data["reliability_issues_low"].sum()}')
print(f'Total Count for Maintainability Issues (Low): {all_data["maintainability_issues_low"].sum()}')
print(f'Total Count for Total Debt (Low): {all_data["total_debt_low"].sum()}')

print(f'Total Count for Security Issues (Medium): {all_data["security_issues_medium"].sum()}')
print(f'Total Count for Reliability Issues (Medium): {all_data["reliability_issues_medium"].sum()}')
print(f'Total Count for Maintainability Issues (Medium): {all_data["maintainability_issues_medium"].sum()}')
print(f'Total Count for Total Debt (Medium): {all_data["total_debt_medium"].sum()}')

print(f'Total Count for Security Issues (High): {all_data["security_issues_high"].sum()}')
print(f'Total Count for Reliability Issues (High): {all_data["reliability_issues_high"].sum()}')
print(f'Total Count for Maintainability Issues (High): {all_data["maintainability_issues_high"].sum()}')
print(f'Total Count for Total Debt (High): {all_data["total_debt_high"].sum()}')
```

```python
# Plotting
fig, axes = plt.subplots(nrows=3, figsize=(12, 18))

# Bar plot for Security Issues (Low)
axes[0].bar(all_data['version'], all_data['security_issues_low'], label='Security Issues (Low)', color='green')
axes[0].set_title('Total Count of Security Issues (Low) Over Versions')
axes[0].set_xlabel('Version')
axes[0].set_ylabel('Total Count (Low)')
axes[0].legend()

# Bar plot for Security Issues (Medium)
axes[1].bar(all_data['version'], all_data['security_issues_medium'], label='Security Issues (Medium)', color='orange')
axes[1].set_title('Total Count of Security Issues (Medium) Over Versions')
axes[1].set_xlabel('Version')
axes[1].set_ylabel('Total Count (Medium)')
axes[1].legend()

# Bar plot for Security Issues (High)
axes[2].bar(all_data['version'], all_data['security_issues_high'], label='Security Issues (High)', color='red')
axes[2].set_title('Total Count of Security Issues (High) Over Versions')
axes[2].set_xlabel('Version')
axes[2].set_ylabel('Total Count (High)')
axes[2].legend()

plt.tight_layout()
plt.show()
```

```python
# Calculate mean and standard deviation for Maintainability Issues (Low)
mean_low = all_data['maintainability_issues_low'].mean()
std_low = all_data['maintainability_issues_low'].std()

# Define control limits (e.g., 3 standard deviations from the mean)
upper_limit_low = mean_low + 3 * std_low
lower_limit_low = mean_low - 3 * std_low

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))

# Line plot for Maintainability Issues (Low)
ax.plot(all_data['version'], all_data['maintainability_issues_low'], label='Maintainability Issues (Low)', color='green', marker='o')

# Plot upper and lower control limits
ax.axhline(y=upper_limit_low, color='red', linestyle='--', label='Upper Control Limit')
ax.axhline(y=lower_limit_low, color='blue', linestyle='--', label='Lower Control Limit')

ax.set_title('Control Chart for Maintainability Issues (Low) Over Versions')
ax.set_xlabel('Version')
ax.set_ylabel('Maintainability Issues (Low)')
ax.legend()

plt.tight_layout()
plt.show()
```
