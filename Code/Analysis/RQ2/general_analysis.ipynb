```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Count the non-null values
counts = {
    'React': df_combined['version_of_react'].count(),
    'Vue': df_combined['version_of_vue'].count(),
    'Svelte': df_combined['version_of_svelte'].count()
}

# Create a DataFrame from the counts
df_counts = pd.DataFrame(list(counts.items()), columns=['Version', 'Count'])

# Create the countplot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Version', y='Count', data=df_counts, palette='Set2')

# Add count numbers over each bar with space
for i, row in df_counts.iterrows():
    ax.text(i, row.Count + 0.01 * df_counts['Count'].max(), row.Count, color='black', ha="center")

plt.title('Number of releases for each framework')
plt.show()
```
