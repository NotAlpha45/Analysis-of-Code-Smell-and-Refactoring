```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_react_adapt = pd.read_csv('../../../Datasets/RQ2/react/react_adaptable_issues.csv')
df_react_consis = pd.read_csv('../../../Datasets/RQ2/react/react_consistent_issues.csv')
df_react_intentional = pd.read_csv('../../../Datasets/RQ2/react/react_intentional_issues.csv')
```

```python
df_react = process_dataframes('react', df_react_adapt, df_react_consis, df_react_intentional)
```

```python
df_vue_adapt = pd.read_csv('../../../Datasets/RQ2/vue/vue_adaptable_issues.csv')
df_vue_consis = pd.read_csv('../../../Datasets/RQ2/vue/vue_consistent_issues.csv')
df_vue_intentional = pd.read_csv('../../../Datasets/RQ2/vue/vue_intentional_issues.csv')
```

```python
df_vue = process_dataframes('vue', df_vue_adapt, df_vue_consis, df_vue_intentional)
```

```python
df_vue.head()
```

```python
df_combined = pd.concat([df_react, df_vue], axis=1)
df_combined['date_of_react']
```

```python
df_combined.columns
```

```python
columns_for_controlChart_of_react = ['adaptability_total_debt_of_react', 'consistency_total_debt_low_of_react', 'intentionality_total_debt_of_react']

columns_for_controlChart_of_vue = ['adaptability_total_debt_of_vue', 'consistency_total_debt_low_of_vue', 'intentionality_total_debt_of_vue']
```

```python
combined_columns = list(zip(columns_for_controlChart_of_react, columns_for_controlChart_of_vue))
```

```python
fig, axs = plt.subplots(3, 2, figsize=(15, 15))

# Create line plots
for i, (react_col, vue_col) in enumerate(combined_columns):
    # Calculate mean and standard deviation for each column
    react_mean = df_combined[react_col].mean()
    react_std = df_combined[react_col].std()
    vue_mean = df_combined[vue_col].mean()
    vue_std = df_combined[vue_col].std()

    # Calculate LCL and UCL for each column
    react_lcl = react_mean - 3*react_std
    react_ucl = react_mean + 3*react_std
    vue_lcl = vue_mean - 3*vue_std
    vue_ucl = vue_mean + 3*vue_std

    # Plot the data
    sns.lineplot(x='version_of_react', y=react_col, data=df_combined, ax=axs[i, 0], marker ='o')
    axs[i, 0].axhline(react_lcl, color='b', linestyle='--')
    axs[i, 0].axhline(react_ucl, color='r', linestyle='--')
    axs[i, 0].axhline(react_mean, color='g', linestyle='--')
    axs[i, 0].set_title(react_col)
    axs[i, 0].set_xticklabels(axs[i, 0].get_xticklabels(), rotation=90)

    sns.lineplot(x='version_of_vue', y=vue_col, data=df_combined, ax=axs[i, 1], marker ='o')
    axs[i, 1].axhline(vue_lcl, color='b', linestyle='--')
    axs[i, 1].axhline(vue_ucl, color='r', linestyle='--')
    axs[i, 1].axhline(vue_mean, color='g', linestyle='--')
    axs[i, 1].set_title(vue_col)
    axs[i, 1].set_xticklabels(axs[i, 1].get_xticklabels(), rotation=90)

plt.tight_layout()
plt.show()
```
